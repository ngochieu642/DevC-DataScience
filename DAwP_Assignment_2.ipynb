**How many people have the word Chief in their job title?**

```python
import re # Regular Expression
import functools

def haveWord(x):
    # This function return 1 if input has the world "Chief"
    return 1 if len(re.findall("Chief",x)) > 0 else 0

print("No. of people who have 'Chief' in there Job title: ", functools.reduce(lambda a,b: a + haveWord(b), sal['JobTitle'], 0))
```

However, there might be cases where 'Chief' -> ['chief', 'CHIEF','chiEf etc..]. Let's check this out

```python
def haveWord_2(x):
    # This function return 1 if input has the world "Chief" 
    return 1 if len(re.findall("Chief",x, flags=re.IGNORECASE)) > 0 else 0

print("No. of people who seems to have 'Chief' in their name under any format: ", functools.reduce(lambda a,b: a + haveWord_2(b), sal['JobTitle'], 0))
```

#### Let's check this Again by looping through job titles

```python
count = 1
for item in sal['JobTitle']:
    if (haveWord(item) and haveWord_2(item)):
        print('Appear in both:', item)
    elif (haveWord_2(item)):
        print('------>>>---------Only appear in haveWord_2: ',item)
```

**Fun Fact: Is there a correlation between length of the Job Title string and Salary?**

*Hint: corr()*

Add new column to sal: JobLength

```python
# Get a list of length JobTitle
lengthJobTitleList = list(map(lambda x: len(x), sal['JobTitle']))
sal['JobLength'] = lengthJobTitleList
sal
```

```python
# Find the correlation
plt.rcParams["figure.figsize"] = (15,7)
sns.jointplot(x='JobLength', y='TotalPayBenefits', data=sal, kind="kde")
```

- **Perfect:** If the value is near ± 1, then it said to be a perfect correlation: as one variable increases, the other variable tends to also increase (if positive) or decrease (if negative).
- **High degree**: If the coefficient value lies between ± 0.50 and ± 1, then it is said to be a strong correlation.
- **Moderate degree**: If the value lies between ± 0.30 and ± 0.49, then it is said to be a medium correlation.
- **Low degree**: When the value lies below ± .29, then it is said to be a small correlation.
No correlation: When the value is zero.

## Pearson Coefficient
![](https://www.statisticssolutions.com/wp-content/uploads/2019/09/ehtsht.png)

```python
sal['TotalPayBenefits'].corr(sal['JobLength'], method="pearson")
```

## Kendall Coefficient

- Kendall rank correlation is a non-parametric test that measures the strength of dependence between two variables
- If we consider two samples, a and b, where each sample size is n, we know that the total number of pairings with a b is n(n-1)/2.  The following formula is used to calculate the value of Kendall rank correlation:

![](http://www.statisticssolutions.com/wp-content/uploads/2010/01/t.jpg)
- Nc= number of concordant
- Nd= Number of discordant

```python
sal['TotalPayBenefits'].corr(sal['JobLength'], method="kendall")
```

## Spearman Coefficient
![](http://www.statisticssolutions.com/wp-content/uploads/2010/01/wher.jpg)
- ρ= Spearman rank correlation
- di= the difference between the ranks of corresponding variables
- n= number of observations

```python
sal['TotalPayBenefits'].corr(sal['JobLength'], method="spearman")
```

**Conclusion** : Low Degree correlation, almost no correlation
